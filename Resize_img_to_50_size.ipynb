```python
from bing_image_downloader import downloader
import cv2
import glob
import os
```

```python
downloader.download("cricketer", limit=100,adult_filter_off=True, force_replace=False, timeout=60, verbose=True)
```

```python
def resize_image(image_path, output_path):
    img = cv2.imread(image_path)
    resized_img = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
    cv2.imwrite(output_path, resized_img)
```

```python
image_files = [f for f in os.listdir('/content/dataset/cricketer') if f.endswith(".jpg")]
for image_file in image_files:
    image_path = os.path.join('/content/dataset/cricketer', image_file)
    output_path = os.path.join('/content/dataset/resized images', image_file)
    resize_image(image_path, output_path)
```
